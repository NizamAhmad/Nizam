```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading the training file
NL1= pd.read_csv('DengAI_Predicting_Disease_Spread_-_Training_Data_Features.csv')
```

```python
#Importing Total number of cases from the Total cases file
NL2= pd.read_csv('DengAI_Predicting_Disease_Spread_-_Training_Data_Labels.csv')
NL2.head(10)
```

```python
# Importing the Total Number of cases 
NL1['Total_Cases'] = NL2['total_cases']
NL1.head(10)
```

```python
NL1 = NL1.fillna(value=NL1.mean())
NL1.head(5)
```

```python
x = NL1[[ 'weekofyear',  'reanalysis_air_temp_k',
       'reanalysis_dew_point_temp_k',
       'reanalysis_max_air_temp_k', 'reanalysis_min_air_temp_k',
       
       'reanalysis_relative_humidity_percent', 'reanalysis_sat_precip_amt_mm',
       'reanalysis_specific_humidity_g_per_kg', 
      'station_max_temp_c',
       'station_min_temp_c', ]]
```

```python
y = NL1['Total_Cases'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 380)
```

```python
# Importing Polynomial regression function from SKlearn
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x_train)
poly.fit(x_train,y_train)
```

```python
#importing Linear regression to run with x_poly from
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_poly, y_train)
```

```python
# Predicting Total Cases
y_pred = model.predict(poly.fit_transform(x_test))
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test, y_pred)
```

```python
metrics.mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
sns.distplot(y_test-y_pred)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(y_test, y_pred)
R2
```

```python
model.score(x_poly, y_train)
```
